```python
# Import packages needed 
from pathlib import Path
import openpyxl

# Pulling the data sheets from the dir. and sorting them
input_dir = Path.cwd()
excel = list(input_dir.rglob("*.xls*"))
excel.sort()

# Iterating through each file and pulls the H2 value
fileNameList = []
H2ValueList = []
for excelData in excel:
    fileName = (Path(excelData).stem)
    fileNameList.append(fileName)
    wb = openpyxl.load_workbook(excelData)
    sheet = wb['Results']
    H2Value = sheet['H2'].value
    H2ValueList.append(H2Value)

# Creating the Workbook, sheets, and naming them
fileNameListSize = len(fileNameList)
wb = openpyxl.Workbook()
sheetNameList = []
count = 0
for Num in range(fileNameListSize):
    wordbreak = fileNameList[Num].split("-")
    s = '-'.join(wordbreak[1:-3])
    if s not in sheetNameList:
        wb.create_sheet(index = count, title = s)
        count += 1
        sheetNameList.append(s)

# adding the fix data to the sheet
R2 = 2
R3 = 3
R4 = 4
R5 = 5
R6 = 6
for Num in range(fileNameListSize):
    wordbreak = fileNameList[Num].split("-")
    current = '-'.join(wordbreak[1:-3])
    
    wordbreak = fileNameList[Num - 1].split("-")
    past = '-'.join(wordbreak[1:-3])
    
    wordbreak = fileNameList[Num].split("-")
    sheetName = '-'.join(wordbreak[1:-3])
    sheet = wb[sheetName]

    if current != past:
        R2 = 2
        R3 = 3
        R4 = 4
        R5 = 5
        R6 = 6 
    
    wordbreak = fileNameList[Num].split("-")
    current2 = '-'.join(wordbreak[:-2])
    
    wordbreak = fileNameList[Num - 1].split("-")
    past2 = '-'.join(wordbreak[:-2])

    if current2 != past2:
        sheet.cell(row = R2, column = 2).value = "slide(sd)"
        sheet.cell(row = R2, column = 3).value = "slice(sc)"
        sheet.cell(row = R2, column = 4).value = "Q¯ [-]"
        sheet.cell(row = R2, column = 5).value = "P [-]"
        sheet.cell(row = R2, column = 6).value = "A,frame[mm²]"
        sheet.cell(row = R2, column = 7).value = " H,disector[µm]"
        sheet.cell(row = R2, column = 8).value = "cell density (cells/mm3)"
        sheet.cell(row = R2, column = 10).value = "volume (mm3)"
        sheet.cell(row = R2, column = 12).value = "cell numbers/slice"
        sheet.cell(row = R2, column = 13).value = "slices represented"
        sheet.cell(row = R2, column = 14).value = "cell number/20slices"
        sheet.cell(row = R2, column = 15).value = "volume (mm3)"
        sheet.cell(row = R2, column = 16).value = "file name"
        sheet.cell(row = R2, column = 19).value = "height of disector (um)"
        sheet.cell(row = R2, column = 20).value = "ROI"

        sheet.cell(row = R3, column = 2).value = "SD1"
        sheet.cell(row = R4, column = 2).value = "SD2"
        sheet.cell(row = R5, column = 2).value = "SD3"
        sheet.cell(row = R6, column = 2).value = "SD4"

        sheet.cell(row = R3, column = 11).value = 2
        sheet.cell(row = R4, column = 11).value = 2
        sheet.cell(row = R5, column = 11).value = 2
        sheet.cell(row = R6, column = 11).value = 2

        sheet.cell(row = R3, column = 13).value = 20
        sheet.cell(row = R4, column = 13).value = 20
        sheet.cell(row = R5, column = 13).value = 20
        sheet.cell(row = R6, column = 13).value = 20

        R2 += 8
        R3 += 8
        R4 += 8
        R5 += 8
        R6 += 8
  
# adding the formulas to the sheets
R3 = 3
R4 = 4
R5 = 5
R6 = 6
R7 = 7 
for Num in range(fileNameListSize):
    wordbreak = fileNameList[Num].split("-")
    current = '-'.join(wordbreak[1:-3])

    wordbreak = fileNameList[Num - 1].split("-")
    past = '-'.join(wordbreak[1:-3])
    
    wordbreak = fileNameList[Num].split("-")
    sheetName = '-'.join(wordbreak[1:-3])
    sheet = wb[sheetName]
    if current != past:
            R2 = 2
            R3 = 3
            R4 = 4
            R5 = 5
            R6 = 6 
            R7 = 7 
             
    wordbreak = fileNameList[Num].split("-")
    current2 = '-'.join(wordbreak[:-2])
    
    wordbreak = fileNameList[Num - 1].split("-")
    past2 = '-'.join(wordbreak[:-2])

    if current2 != past2:
        stringR3 = str(R3)
        stringR4 = str(R4)
        stringR5 = str(R5)
        stringR6 = str(R6)
        stringR7 = str(R7)
        sheet.cell(row = R7, column = 8).value = "=AVERAGE(H" + stringR3 + ":H" + stringR6 + ")"
        sheet.cell(row = R7, column = 10).value = "=AVERAGE(J" + stringR3 + ":J" + stringR6 + ")"

        sheet.cell(row = R3, column = 12).value = "=H" + stringR3 + "*J" + stringR3 + "*K" + stringR3
        sheet.cell(row = R4, column = 12).value = "=H" + stringR4 + "*J" + stringR4 + "*K" + stringR4
        sheet.cell(row = R5, column = 12).value = "=H" + stringR5 + "*J" + stringR5 + "*K" + stringR5
        sheet.cell(row = R6, column = 12).value = "=H" + stringR6 + "*J" + stringR6 + "*K" + stringR6 

        sheet.cell(row = R3, column = 14).value = "=L" + stringR3 + "*M" + stringR3
        sheet.cell(row = R4, column = 14).value = "=L" + stringR4 + "*M" + stringR4
        sheet.cell(row = R5, column = 14).value = "=L" + stringR4 + "*M" + stringR5
        sheet.cell(row = R6, column = 14).value = "=L" + stringR4 + "*M" + stringR6

        sheet.cell(row = R3, column = 15).value = "=J" + stringR3 + "*K" + stringR3 + "*M" + stringR3
        sheet.cell(row = R4, column = 15).value = "=J" + stringR4 + "*K" + stringR4 + "*M" + stringR4
        sheet.cell(row = R5, column = 15).value = "=J" + stringR5 + "*K" + stringR5 + "*M" + stringR5
        sheet.cell(row = R6, column = 15).value = "=J" + stringR6 + "*K" + stringR6 + "*M" + stringR6

        sheet.cell(row = R7, column = 14).value = "=SUM(N" + stringR3 + ":N" + stringR6 + ")"
        sheet.cell(row = R7, column = 15).value = "=SUM(O" + stringR3 + ":O" + stringR6 + ")"

        R3 += 8
        R4 += 8
        R5 += 8
        R6 += 8
        R7 += 8
#adding animal region codes to the sheet
R1 = 1
R1List = []
animalCodeList = []

for Num in range(fileNameListSize):

    wordbreak = fileNameList[Num].split("-")
    current = '-'.join(wordbreak[1:-3])

    wordbreak = fileNameList[Num - 1].split("-")
    past = '-'.join(wordbreak[1:-3])

    wordbreak = fileNameList[Num].split("-")
    sheetName = '-'.join(wordbreak[1:-3])
    sheet = wb[sheetName]

    wordbreak = fileNameList[Num].split("-")
    animalCode = '-'.join(wordbreak[:-2])
    
    if current != past:
        R1 = 1
    if animalCode not in animalCodeList:
        sheet.cell(row = R1, column = 1).value = animalCode
        R1List.append(R1)
        animalCodeList.append(animalCode)
        R1 += 8

#adding values to sheet
ColNum = 0
Volume = "vol"
sd1Row = 0
sd2Row = 0
sd3Row = 0
sd4Row = 0

for Num in range(fileNameListSize):

    #find the colNum
    wordbreak = fileNameList[Num].split("-")
    VolOrDen = '-'.join(wordbreak[-1:])
    if VolOrDen.casefold() == Volume.casefold():
        ColNum = 10
    else:
        ColNum = 8 

    #locate the sheet
    wordbreak = fileNameList[Num].split("-")
    sheetName = '-'.join(wordbreak[1:-3])
    sheet = wb[sheetName]
    
    # find the row number
    wordbreak = fileNameList[Num].split("-")
    TempAnimalCode = '-'.join(wordbreak[:-2])
    index = animalCodeList.index(TempAnimalCode)
    Location = R1List[index]
    wordbreak = fileNameList[Num].split("-")
    SdNum = '-'.join(wordbreak[-2:-1])

    if SdNum.casefold() == "sd1".casefold():
        sd1Row = (Location + 2)
        sheet.cell(row = sd1Row, column = ColNum).value = H2ValueList[Num]
    elif SdNum.casefold() == "sd2".casefold():
        sd2Row = (Location + 3)
        sheet.cell(row = sd2Row, column = ColNum).value = H2ValueList[Num]
    elif SdNum.casefold() == "sd3".casefold():
        sd3Row = (Location + 4)
        sheet.cell(row = sd3Row, column = ColNum).value = H2ValueList[Num]
    elif SdNum.casefold() == "sd4".casefold():
        sd4Row = (Location + 5)
        sheet.cell(row = sd4Row, column = ColNum).value = H2ValueList[Num]

# saving the excel
excelTitle = input("Enter the title of the excel file here:")
wb.save(excelTitle + ".xlsx")
```
